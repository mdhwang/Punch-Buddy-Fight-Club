```python
import pandas as pd
from clean_data import known_clean_data

import warnings
warnings.filterwarnings('ignore')

import pickle
```

```python
accel_jab = pd.read_csv('data/25hz-Accel-Gyro/Left Jab 50 Matt_Accelerometer.csv')
gyro_jab = pd.read_csv('data/25hz-Accel-Gyro/Left Jab 50 Matt_Gyroscope.csv')
accel_hook = pd.read_csv('data/25hz-Accel-Gyro/Left Hook 50 Matt_Accelerometer.csv')
gyro_hook = pd.read_csv('data/25hz-Accel-Gyro/Left Hook 50 Matt_Gyroscope.csv')
accel_cross = pd.read_csv('data/25hz-Accel-Gyro/Right Cross 50 Matt_Accelerometer.csv')
gyro_cross = pd.read_csv('data/25hz-Accel-Gyro/Right Cross 50 Matt_Gyroscope.csv')
```

```python
cleaned_jab = known_clean_data(accel_jab,gyro_jab,50,8,"Left Jab")
cleaned_hook = known_clean_data(accel_hook,gyro_hook,50,8,"Left Hook")
cleaned_cross = known_clean_data(accel_cross,gyro_cross,50,8,"Right Cross")
```

```python
pickle.dump(cleaned_jab, open('50_left_jab.p','wb'))
unjar = pickle.load(open('50_left_jab.p','rb'))
```

```python
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
plt.figure(figsize=(16, 6))
for each in cleaned_jab:
    sb.lineplot(each["time"],each["magnitude"],color='b',alpha=0.25)
    plt.title("50x Jabs - Combined Magnitude",fontsize=20)
plt.figure(figsize=(16, 6))
for each in cleaned_hook:
    sb.lineplot(each["time"],each["magnitude"],color='r',alpha=0.25)
    plt.title("50x Hooks - Combined Magnitude",fontsize=20)
plt.figure(figsize=(16, 6))
for each in cleaned_cross:
    sb.lineplot(each["time"],each["magnitude"],color='g',alpha=0.25)
    plt.title("50x Crosses - Combined Magnitude",fontsize=20)
```

```python
plt.figure(figsize=(16, 6))
for each in cleaned_jab:
    sb.lineplot(each["time"],each["X-Axis (°/s)"],color='b',alpha=0.25)
    sb.lineplot(each["time"],each["Y-Axis (°/s)"],color='r',alpha=0.25)
    sb.lineplot(each["time"],each["Z-Axis (°/s)"],color='g',alpha=0.25)
    plt.title("50x Jabs - Angular Velocity",fontsize=20)
plt.figure(figsize=(16, 6))
for each in cleaned_hook:
    sb.lineplot(each["time"],each["X-Axis (°/s)"],color='b',alpha=0.25)
    sb.lineplot(each["time"],each["Y-Axis (°/s)"],color='r',alpha=0.25)
    sb.lineplot(each["time"],each["Z-Axis (°/s)"],color='g',alpha=0.25)
    plt.title("50x Hooks - Angular Velocity",fontsize=20)
plt.figure(figsize=(16, 6))
for each in cleaned_cross:
    sb.lineplot(each["time"],each["X-Axis (°/s)"],color='b',alpha=0.25)
    sb.lineplot(each["time"],each["Y-Axis (°/s)"],color='r',alpha=0.25)
    sb.lineplot(each["time"],each["Z-Axis (°/s)"],color='g',alpha=0.25)
    plt.title("50x Crosses - Angular Velocity",fontsize=20)
```
